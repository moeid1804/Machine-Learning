# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split,cross_val_score,cross_validate
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```python
df=pd.read_csv('boston.csv')
df.head()
```

## Explore Data

```python
df.isnull().sum()
```

```python
correlation=df.corr()
sns.heatmap(correlation,annot=True,cmap='Greens',fmt='.1f')
```

```python
df.hist(figsize=(13,12))
plt.show()
```

## Data preprocessing

```python
df.drop(['Unnamed: 0','DIS','CHAS','B'],axis=1,inplace=True)
df.head()
```

```python
x=df.drop(['Price'],axis=1)
y=df['Price']
```

```python
x
```

## Create the model

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3)
```

```python
model=XGBRegressor()
model.fit(x_train,y_train)
pred=model.predict(x_test)
```

## Evaluate model

```python
metric_1=r2_score(y_test,pred)
print('The model successful to predict price with accuracy',metric_1)
metric_2=mean_absolute_error(y_test,pred)
print('The mean absolute error is ',metric_2)
```

# Difference between Actual and Predicted Price

```python
plt.scatter(y_train,model.predict(x_train),color='r')
plt.title('Destribution of Training & Testing')
plt.xlabel('Actual Data')
plt.ylabel('Predicting Data')
plt.scatter(y_test,pred)
plt.show()
```
